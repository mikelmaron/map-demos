```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv('cities_for_map.csv')
```

```python
df_2022 = df[df['Year'] == 2022] 
```

```python
gdf_2022 = gpd.GeoDataFrame(df_2022, geometry=gpd.GeoSeries.from_xy(df['Longitude'], df['Latitude']), crs=4326)
```

```python
columns = {
    'City': 'City',
    'Country': 'Country',
    'Country ISO3': 'ISO3',
    'geometry': 'geometry',
    'Population': 'Pop',
    'Emissions: Total (tons CH4/year)': 'Emissions Total',
    'Percentile: Emissions per Capita (%)': 'Emissions Capita Percentile',
    'Waste Generation Rate (tons/year)': 'waste annual',
    'Waste Components: Food (%)' : 'Food %',
    'Waste Components: Green (%)' : 'Green %',
    'Waste Components: Wood (%)': 'Wood %',
    'Waste Components: Paper and Cardboard (%)': 'Paper %',
    'Waste Components: Textiles (%)': 'Textiles %',
    'Waste Components: Plastic (%)': 'Plastic %',
    'Waste Components: Metal (%)': 'Metal %', 
    'Waste Components: Glass (%)': 'Glass %',
    'Waste Components: Rubber/Leather (%)': 'Rubber %',
    'Waste Components: Other (%)': 'Other %',
    'Fraction of Waste to Methane Capture Landfills (%)': 'Capture %',
    'Methane Capture Efficiency (%)': 'Capture Efficiency %',
    'Diversons: Compost (%)': 'Compost %',
    'Diversons: Anaerobic Digestion (%)': 'Digestion %',
    'Diversons: Incineration (%)': 'Incineration %',
    'Diversons: Recycling (%)': 'Recycling %'
}

for col in gdf_2022.columns[1:]:
    if col not in columns.keys():
        gdf_2022.drop(col, axis=1, inplace=True)

gdf_2022.rename(columns, axis=1, inplace=True)
```
